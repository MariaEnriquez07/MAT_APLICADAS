```python
import numpy as np
import matplotlib.pyplot as plt
temperaturas = [25.6, 26.1, 24.8, 27.3, 28.4, 26.7, 25.9, 29.1, 30.2, 27.8, 24.3, 23.9, 26.5, 28.0, 27.4, 29.3, 30.5, 26.8, 25.7, 24.6]
media = np.mean(temperaturas)
mediana = np.median(temperaturas)
desviacion = np.std(temperaturas)

plt.figure(figsize=(8,5))
plt.plot(range(1, len(temperaturas)+1), temperaturas, marker='o', linestyle='-', color='b', label='Temperatura mensual')
plt.axhline(media, color='r', linestyle='--', label=f'Media: {media:.2f}')
plt.xlabel('Mes')
plt.ylabel('Temperatura (°C)')
plt.title('Temperatura media mensual')
plt.legend()
plt.show()
```

```python
g = 9.81
v0 = float(input("Ingrese la velocidad inicial (m/s): "))
theta = float(input("Ingrese el ángulo de lanzamiento (grados): "))
theta = np.radians(theta)
t_flight = (2 * v0 * np.sin(theta)) / g
t = np.linspace(0, t_flight, num=100)
x = v0 * np.cos(theta) * t
y = v0 * np.sin(theta) * t - 0.5 * g * t**2
plt.plot(x, y, '-r')
plt.title("Movimiento Parabólico")
plt.xlabel("Distancia (m)")
plt.ylabel("Altura (m)")
plt.grid()
plt.show()
```

```python
def sierpinski(order, points):
    if order == 0:
        plt.fill(points[:, 0], points[:, 1], 'k')
    else:
        mid1 = (points[0] + points[1]) / 2
        mid2 = (points[1] + points[2]) / 2
        mid3 = (points[2] + points[0]) / 2
        sierpinski(order-1, np.array([points[0], mid1, mid3]))
        sierpinski(order-1, np.array([mid1, points[1], mid2]))
        sierpinski(order-1, np.array([mid3, mid2, points[2]]))

orden = int(input("Ingrese el nivel del fractal de Sierpinski (máx 6): "))
if orden > 6:
    print("Nivel demasiado alto, cambiando a 6 para evitar bloqueos.")
    orden = 6

puntos = np.array([[0, 0], [1, np.sqrt(3)/2], [2, 0]])
plt.figure(figsize=(5, 5))
sierpinski(orden, puntos)
plt.axis('off')
plt.show()
```
