La fórmula del volumen del cilindro es:

$V = \pi r^2h$

- Paso 3: Relacion entre variables

Como el cilindro esta inscrito dentro del cono, existe una relacion entre $\frac{r}{R} = \frac{H - h}{H} \Rightarrow r = R\left(1 - \frac{h}{H}\right)$

- Paso 4: Derivada y puntos criticos

$V(h) = \pi R^2\left(1 - \frac{h}{H}\right)^2$

$V(h) = \pi R^2\left[\left(1 - \frac{h}{H}\right)^2 - 2\left(1 - \frac{h}{H}\right)\left(\frac{h}{H}\right) \right]$

Factorizamos:

$V(h) = \pi R^2\left(1 - \frac{4h}{H} + \frac{3h^2}{H^2}\right)$

Igualar a 0:

$1 - \frac{4h}{H} + \frac{3h^2}{H^2} = 0$

Multiplicar todo por $H^2$

$H^2 - AhH + 3h^2 = 0$

Formula cuadratica:

$h = H\quad \text{o} \quad h = \frac{H}{3}$

- Paso 5: Evaluacion

> Si $\left( h = H\right)$, entonces $\left( r = 0\Rightarrow V = 0 \right)$


>Si $\left( h = \frac{H}{3}\right)$ entonces:

$r = R\left(1 - \frac{1}{3}\right) = \frac{2R}{3}$

Este valor da el **valomen maximo**


- Conclusion
El cilindro de volumen maximoo inscrito en el cono tiene:

```python
def metodo_newton(f, df, x0, tol=1e-4, max_iter=100):
    x = x0
    for i in range(max_iter):
        fx = f(x)
        dfx = df(x)
        if dfx == 0:
            break
        x_new = x - fx / dfx
        if abs(x_new - x) < tol:
            return x_new
        x = x_new
    return x

f1 = lambda x: x**5 - 6*x**3 + 2*x - 2
df1 = lambda x: 5*x**4 - 18*x**2 + 2

x_vals = np.linspace(-3, 3, 400)
plt.plot(x_vals, f1(x_vals), label='f(x) = x^5 - 6x^3 + 2x - 2')
plt.axhline(0, color='black', linestyle='--')
plt.title("Gráfica de f(x) = x^5 - 6x^3 + 2x - 2")
plt.grid()
plt.legend()
plt.show()

x0_vals_f1 = [-2.2, -0.3, 1.8]
for i, x0 in enumerate(x0_vals_f1, start=1):
    raiz = metodo_newton(f1, df1, x0)
    print(f"Raíz aproximada {i} : {raiz}")
```

1.
Analicemos la función:  
$f(x) = \frac{x}{x + 1}$

1.
Dominio: $x \in (-\infty, -1) \cup (-1, \infty)$

2. Intersección con los ejes:
- Eje $y$: $f(0) = 0$ ⇒ punto $(0, 0)$
- Eje $x$: $f(x) = 0$ cuando el numerador es cero ⇒ $x = 0$

3. Asíntotas:
- Vertical: $x = -1$
- Horizontal:
  $\lim_{x \to \pm \infty} \frac{x}{x+1} = 1$
  Asíntota horizontal: $y = 1$

4. Derivada:
$
f'(x) = \frac{(1)(x+1) - x(1)}{(x+1)^2} = \frac{1}{(x+1)^2}
$


5. Segunda derivada:
$f''(x) = \frac{-2}{(x+1)^3}$
- $f''(x) < 0$ si $x > -1$ ⇒ cóncava hacia abajo
- $f''(x) > 0$ si $x < -1$ ⇒ cóncava hacia arriba
- Hay un punto de inflexión en $x = -1$ aunque no está en el dominio.

La gráfica muestra una asíntota vertical en $x = -1$ y una horizontal en $y = 1$. La función es creciente en todo s

```python
def f(x):
    return x / (x + 1)

x_vals = np.linspace(-5, 5, 1000)
x_vals = x_vals[np.abs(x_vals + 1) > 0.01]

plt.figure(figsize=(10, 6))
plt.plot(x_vals, f(x_vals), label=r'$f(x) = \frac{x}{x+1}$', color='blue')

plt.axvline(x=-1, color='red', linestyle='--', label='Asíntota vertical $x = -1$')
plt.axhline(y=1, color='green', linestyle=':', label='Asíntota horizontal $y = 1$')

plt.axhline(y=0, color='black', linewidth=0.5)
plt.axvline(x=0, color='black', linewidth=0.5)
plt.title('Gráfica de $f(x) = \\frac{x}{x+1}$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.grid(True)
plt.ylim(-10, 10)
plt.legend()
plt.tight_layout()

plt.figtext(0.5, -0.05, "Figura 1: Gráfica de $f(x) = \\frac{x}{x+1}$ con asíntotas vertical y horizontal.", ha='center', fontsize=10)

plt.show()
```

2.
Analicemos la función:  
$g(x) = \frac{|x| + x}{2}(3x + 2)$

1. Dominio: $\mathbb{R}$

2. Simplificación por tramos:

Recordemos que:
$\frac{|x| + x}{2} =
\begin{cases}
0 & \text{si } x < 0 \\
x & \text{si } x \ge 0
\end{cases}
$

Entonces:
$g(x) =
\begin{cases}
0 & \text{si } x < 0 \\
x(3x + 2) = 3x^2 + 2x & \text{si } x \ge 0
\end{cases}
$

3. Interseccione con los ejes:

- Eje $y$: $g(0) = 0$
- Eje $x$:  
  Resolver $g(x) = 0$:
   Para $x < 0$, $g(x) = 0$  
   Para $x \ge 0$:  
    $$3x^2 + 2x = 0 \Rightarrow x(3x + 2) = 0 \Rightarrow x = 0, \, x = -\frac{2}{3}$$  
    Pero $x = -\frac{2}{3}$ no pertenece al tramo $x \ge 0$, así que no aplica

Resultado:  
$g(x) = 0 \text{ en todo } x < 0 \text{ y en } x = 0$


4. Derivada por tramos:

$g'(x) =
\begin{cases}
0 & \text{si } x < 0 \\
(3x^2 + 2x)' = 6x + 2 & \text{si } x > 0
\end{cases}$

no existe deriada en $x = 0$ porque hay un cambio brusco de pendiente (punto anguloso)

5. Segunda derivada:

$g''(x) =
\begin{cases}
0 & \text{si } x < 0 \\
(6x + 2)' = 6 & \text{si } x > 0
\end{cases}
$

Conclusión:  
- La función es constante e igual a 0 para $x < 0$  
- A partir de $x = 0$, crece como un trinomio cuadrático  
- No es derivable en $x = 0$  
- Tiene una concavidad hacia arriba para $x > 0$

```python
def g(x):
    return np.where(x < 0, 0, x * (3 * x + 2))

x_vals = np.linspace(-5, 5, 1000)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, g(x_vals), label=r'$g(x) = \frac{|x| + x}{2}(3x + 2)$', color='purple')

plt.axhline(y=0, color='black', linewidth=0.5)
plt.axvline(x=0, color='black', linewidth=0.5)
plt.title('Gráfica de $g(x) = \\frac{|x| + x}{2}(3x + 2)$')
plt.xlabel('$x$')
plt.ylabel('$g(x)$')
plt.grid(True)
plt.legend()

plt.figtext(0.5, -0.05, "Figura 2: Gráfica de $g(x) = \\frac{|x| + x}{2}(3x + 2)$, función por tramos con cambio en $x = 0$.", ha='center', fontsize=10)

plt.show()
```

```python
x = np.linspace(0, 10, 400)

# Función cúbica con inflexión en x=6 y cóncava hacia arriba antes de 6
def f(x):
    return (x - 6)**3 + 3*(x - 6) + 5

plt.figure(figsize=(8, 5))
plt.plot(x, f(x), label="f(x)", color='green')
plt.axvline(x=6, linestyle='--', color='gray', label='x = 6 (Inflexión)')
plt.title("f''(6) = 0 y f''(x) > 0 para 2 < x < 6")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
